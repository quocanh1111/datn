```python
# Paste this in any Jupyter notebook cell
import sys
import importlib

print(f"Python: {sys.version.split()[0]}")
print(f"Container: {__import__('socket').gethostname()}")
print()

packages = [
    'pyspark', 'pyarrow', 'pandas', 'numpy',
    'boto3', 'minio', 'faker', 'findspark',
    'requests', 'psycopg2', 'airflow',
    'cryptography', 'py4j'
]

print(f"{'Package':<20} {'Status':<10} Version")
print("-" * 45)
for pkg in packages:
    try:
        mod = importlib.import_module(pkg)
        ver = getattr(mod, '__version__', 'unknown')
        print(f"  {pkg:<18} {'OK':<10} {ver}")
    except ImportError as e:
        print(f"  {pkg:<18} {'MISSING':<10} -")

print()
print("Environment:")
import os
for var in ['SPARK_HOME', 'JAVA_HOME', 'PYSPARK_PYTHON', 'AIRFLOW_HOME']:
    print(f"  {var:<20} {os.environ.get(var, 'NOT SET')}")
```
