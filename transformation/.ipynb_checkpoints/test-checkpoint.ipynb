```python
import io
import json
import boto3
from botocore.client import Config

# Connect to MinIO
s3 = boto3.client(
    "s3",
    endpoint_url="http://storage:9000",
    aws_access_key_id="root",
    aws_secret_access_key="password",
    config=Config(signature_version="s3v4"),
    region_name="us-east-1",
)

# Target the exact nested metadata file path inside your bronze bucket
bucket_name = "bronze"
file_key = "metadata-dump/listenbrainz-public-dump-20260415-000003/lbdump/recording_feedback"

print(f"Opening streaming channel to: s3://{bucket_name}/{file_key}...")

try:
    obj = s3.get_object(Bucket=bucket_name, Key=file_key)
    
    print("\n--- METADATA SCHEMATIC BREAKDOWN ---")
    with io.TextIOWrapper(obj["Body"], encoding="utf-8") as file:
        for i in range(2):
            line = file.readline()
            if line:
                record = json.loads(line.strip())
                print(f"\n[Record #{i+1} Keys & Sample Content Type]:")
                for key, val in record.items():
                    if isinstance(val, list):
                        sample_keys = list(val[0].keys()) if len(val) > 0 and isinstance(val[0], dict) else "Primitive Array"
                        print(f" - {key}: List (Length: {len(val)}) -> Sample: {sample_keys}")
                    elif isinstance(val, dict):
                        print(f" - {key}: Dictionary -> Keys: {list(val.keys())}")
                    else:
                        print(f" - {key}: {type(val).__name__} (Sample: {val})")
except Exception as e:
    print(f"Failed to read object. Error details: {e}")
```
